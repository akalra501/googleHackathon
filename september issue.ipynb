```python
import pandas as pd
import numpy as np
```

```python
df2 = pd.read_csv("dataset2b.csv")     # Julâ€“Dec file (the one with September)
df1 = pd.read_csv("dataset1.csv")      # to get is_eu / region

df2["date_dt"] = pd.to_datetime(df2["date"], errors="coerce")
df2 = df2.merge(df1[["customer_id","is_eu","region"]], on="customer_id", how="left")

df2["month"] = df2["date_dt"].dt.to_period("M").astype(str)

# Missingness by month
miss_by_month = df2.groupby("month")[["logins","feature_events","session_minutes"]].apply(lambda g: g.isna().sum())
print(miss_by_month)

# Focus on Sep 2024
sept = df2[df2["month"] == "2024-09"].copy()
print("Sep rows:", sept.shape[0])
print("Sep missing:", sept[["logins","feature_events","session_minutes"]].isna().sum())

# Confirm it's EU-only
print(sept.groupby("is_eu")[["logins","feature_events","session_minutes"]].apply(lambda g: g.isna().mean()))
```
